```python
import numpy as np
import pandas as pd
import os
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from pathlib import Path
from sklearn.pipeline import Pipeline
```

```python
cwd = Path.cwd().parent
```

```python
with open(cwd / 'data/train.csv') as f:
    data = pd.read_csv(f)
data


with open(cwd / 'data/test.csv') as testf:
    test_data = pd.read_csv(testf)
```

```python
data = pd.get_dummies(data)
test_data = pd.get_dummies(test_data)
data.head()
```

```python
test_data.head()
```

```python
dropcolumn = []
dropline = []
for i in data:
    if np.sum(np.isnan(data[i]))/len(data[i]) >=0.1:
        dropcolumn.append(i)
    if np.sum(data[i])==1:
        wh = np.where(data[i]==1)[0]
        dropline.append(wh.tolist()[0])
        dropcolumn.append(i)

print(dropcolumn)
data = data.drop(columns=dropcolumn)
data = data.drop(dropline)
data.head()

#test_data = test_data.drop(columns=dropcolumn)


data = data.fillna(0)
test_data = test_data.fillna(0)
```

```python
x_train = data.drop(columns=['SalePrice'])
y_train = data['SalePrice']
x_test = test_data

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2,random_state=42)
x_test
```

```python
features = test_data.columns.tolist()
x_test = pd.DataFrame(x_test)
x_test = x_test[features]

print(x_train.columns)
print(x_test.columns)
```

```python
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)
x_val = scaler.fit_transform(x_val)
x_test = scaler.fit_transform(x_test)
```

```python
regr = LinearRegression() 
```

```python
regr.fit(x_train, y_train) 
print(regr.score(x_val, y_val)) 
```

```python
regr.predict(x_test)
```
